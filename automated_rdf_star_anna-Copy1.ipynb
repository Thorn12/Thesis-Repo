```python
from rdflib import URIRef, BNode, Literal, Namespace, Graph
from rdflib.namespace import FOAF, DCTERMS, XSD, RDF, SDO
import pprint
```

```python
g = Graph()
g.parse("D:\\Downloads\\AnnaTest.ttl")

for stmt in g:
    pprint.pprint(stmt)
```

```python
#Assign variables to proxy, predicate for proxy and URIReferences for the ease of use of calling its triples for each RDF Star

proxy = URIRef('http://data.biographynet.nl/rdf/Person-99999999')
proxy_predicate = URIRef('http://www.intavia.eu/idm-core/person_proxy_for')
description1 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_01')
description2 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_02')
description3 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_03')
description4 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_04')
description5 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_05')
description6 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_06')
description7 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_07')
```

```python
#List for all the RDF Stars statements
RDF_Star_list = []
```

```python
def extend_RDF_Star_list(description):
    # create a graph for the description
    temp_graph1 = Graph()
    # add all triples with subject of the description to a temporary graph
    temp_graph1 += g.triples((description, None, None))
    for stmt in temp_graph1:
        first_statement = stmt
        subject = stmt[2]
        #make a second a temporary graph with the object of tempgraph1 and make it as the subject to get the predicate and object for the RDF Star triples
        temp_graph2 = Graph()
        temp_graph2 = g.triples((subject, None, None))
        for stmt in temp_graph2:
            predicate, triple_object = stmt[:2]
            RDF_Star_list.extend([first_statement, predicate, triple_object])
```

```python
extend_RDF_Star_list(description1)
extend_RDF_Star_list(description2)
extend_RDF_Star_list(description3)
extend_RDF_Star_list(description4)
extend_RDF_Star_list(description5)
extend_RDF_Star_list(description6)
extend_RDF_Star_list(description7)
```

```python
print(RDF_Star_list)
```

```python
#with the descriptions making triples based on the information from the writers of biographies there is a chance that there will be duplicate triples in the list, this function will eliminate duplicate triples
no_dupe_RDF_Star_list = list(dict.fromkeys(RDF_Star_list))
```

```python
# Open the file in write mode
with open('rdf_star_list.txt', 'w') as file:
    # Iterate over each item in the list
    for item in no_dupe_RDF_Star_list:
        item = str(item)
        # Write each item to the file
        file.write(item + '\n')

# File is automatically closed after exiting the 'with' block
```
