```python
#Assign variables to proxy, predicate for proxy and URIReferences for the ease of use of calling its triples for each named graph

real_person = URIRef('http://data.biographynet.nl/rdf/Person-99999999')
proxy_predicate = URIRef('http://www.intavia.eu/idm-core/person_proxy_for')
description1 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_01')
description2 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_02')
description3 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_03')
description4 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_04')
description5 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_05')
description6 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_06')
description7 = URIRef('http://data.biographynet.nl/rdf/BioParts-99999999_07')
```

```python
def create_named_graph(description, proxy_predicate, real_person):
    graph = Graph()
    # create a graph for a description
    # add all triples with subject that matches the description
    graph += g.triples((description, None, None))
    for stmt in graph:
        subject = stmt[2]
        graph += g.triples((subject, None, None))
    graph.add((description, proxy_predicate, real_person))
    return graph
```

```python
ngraph1 = create_named_graph(description1, proxy_predicate, real_person)
ngraph1.serialize(destination="D:\\named_graph1_Anna.ttl")
ngraph2 = create_named_graph(description2, proxy_predicate, real_person)
ngraph2.serialize(destination="D:\\named_graph2_Anna.ttl")
ngraph3 = create_named_graph(description3, proxy_predicate, real_person)
ngraph3.serialize(destination="D:\\named_graph3_Anna.ttl")
ngraph4 = create_named_graph(description4, proxy_predicate, real_person)
ngraph4.serialize(destination="D:\\named_graph4_Anna.ttl")
ngraph5 = create_named_graph(description5, proxy_predicate, real_person)
ngraph5.serialize(destination="D:\\named_graph5_Anna.ttl")
ngraph6 = create_named_graph(description6, proxy_predicate, real_person)
ngraph6.serialize(destination="D:\\named_graph6_Anna.ttl")
ngraph7 = create_named_graph(description7, proxy_predicate, real_person)
ngraph7.serialize(destination="D:\\named_graph7_Anna.ttl")
```
